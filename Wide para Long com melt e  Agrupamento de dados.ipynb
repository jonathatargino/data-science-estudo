```python
import pandas as pd
```

```python
emissoes_gases = pd.read_excel("dados/seeg10_geral_br_uf.xlsx", sheet_name="GEE Estados")
```

```python
emissoes_gases
```

```python
colunas_info = list(emissoes_gases.loc[:, "Nível 1 - Setor":"Produto"].columns)

colunas_info
```

```python
colunas_emissao = list(emissoes_gases.loc[:, 1970:2021])

colunas_emissao
```

```python
emissoes_gases.melt(id_vars=colunas_info, value_vars=colunas_emissao, var_name="Ano", value_name = "Emissão")
```

```python
emissoes_gases
```

```python
emissoes_por_ano = emissoes_gases.melt(id_vars=colunas_info, value_vars=colunas_emissao, var_name="Ano", value_name = "Emissão")
```

```python
emissoes_por_ano_agrupado_por_gas = emissoes_por_ano.groupby("Gás")
```

```python
emissoes_por_ano_agrupado_por_gas.groups
```

```python
emissoes_por_ano_agrupado_por_gas.get_group("CO2 (t)")
```

```python
emissoes_por_ano_agrupado_por_gas[["Emissão"]].sum()
```

```python
emissao_por_gas = emissoes_por_ano_agrupado_por_gas[["Emissão"]].sum().sort_values("Emissão", ascending=False)

emissao_por_gas
```

```python
emissao_por_gas.plot(kind="barh", figsize=(10,6))
```

```python
emissao_por_gas.iloc[0:9]
```

```python
emissao_C02 = emissao_por_gas.iloc[0:9]
```

```python
porcentagem_C02 = (emissao_C02.sum()/emissao_por_gas.sum()).iloc[0]

round(porcentagem_C02 * 100, 2)
```
