```python
series_booleana_mudanca_terra = dados["Nível 1 - Setor"] == "Mudança de Uso da Terra e Floresta"
series_booleana_amazonas = dados["Estado"] == "AM"

emissoes_mudanca_terra_amazonas = dados[(series_booleana_mudanca_terra) & (series_booleana_amazonas)]
emissoes_mudanca_terra_amazonas
```

```python
series_booleana_agropecuaria = dados["Nível 1 - Setor"] == "Agropecuária"
series_booleana_para = dados["Estado"] == "PA"

emissoes_agropecuaria_do_para = dados[(series_booleana_agropecuaria) & (series_booleana_para)]

emissoes_agropecuaria_do_para
```

```python
emissoes_2021 = emissoes_agropecuaria_do_para[2021]
```
